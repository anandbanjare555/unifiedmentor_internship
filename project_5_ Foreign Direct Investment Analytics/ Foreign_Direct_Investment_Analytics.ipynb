```python
import pandas as pd

# Read the data from the provided link
data = pd.read_csv('FDI data.csv')

# Convert the index 'Sector' into a separate column
data = data.reset_index()

# Save the modified data to a CSV file
data.to_csv('modified_FDI_data.csv', index=False)
```

```python
# Calculate the total FDI for each sector
total_fdi = data.sum(axis=1)

# Calculate the percentage of total FDI for each sector
percentage_fdi = total_fdi / total_fdi.sum() * 100

# Plot the percentage of total FDI for each sector
percentage_fdi.plot(kind='bar',figsize=(15,15))
plt.title('Percentage of Total FDI by Sector')
plt.xlabel('Sector')
plt.ylabel('Percentage of Total FDI')
plt.show()
```

```python


# Calculate the correlation between FDI in each sector and time
correlation = data.corr()

# Plot the correlation between FDI in each sector and time
plt.figure(figsize=(50, 100))  # Adjust the figure size for better visualization
plt.matshow(correlation)
plt.title('Correlation between FDI in each sector and time')
plt.xlabel('Year')
plt.ylabel('Sector')
plt.show()
```

```python
# Calculate the total FDI for each year
yearly_total_fdi = data.sum(axis=0)

# Plot the total FDI for each year
yearly_total_fdi.plot(kind='line',figsize=(20,15))
plt.title('Total FDI by Year')
plt.xlabel('Year')
plt.ylabel('Total FDI')
plt.show()
```
